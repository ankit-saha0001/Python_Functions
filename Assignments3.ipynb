### Task 1:- (Price after Discount)
#### Make a function that tales input of price and discount and return discounted price with the default discount value of 5% 

```python
## The function

def discounted_price(price, discount=5):
    ## Check to ensure the discount never incrreases more than 60 percent
    if(discount<=60):
        return (price-(price*discount)/100)
    else :
        return (f"Please enter a valid discount percentage which is less than 60")
        


```

```python
## Testing the function
print(discounted_price(1000,10))
```

```python
print(discounted_price(500))
```

```python
print(discounted_price(10,61))
```

### Task 2 :- (Factorial utility)
#### Creating a recursive fucntion to handle the factorial


```python
## The recursive factorial function
def factorial(n):
    if(n<0):
        return "Please enter a non-negative integer"
    elif(n==1 or n==0):
        return 1
    else: 
        return n*factorial(n-1)
```

```python
## Testing the function
print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

### Task 6:- Combined Utility Function
#### Using the function to apply discount and filter to keep the discounted items prices above 300

```python
## Required function to give discount
discount= lambda x: 0.9*x
```

```python
## Using map function with test price list
process_prices=[100,500,900,50,750]

print(f"THe price before discount is: \n {process_prices}")

discounted_prices=list(map(discount,process_prices))

print(f"The price after the discount is: \n {discounted_prices} ")
```

```python
## Filtering prices above 300
filtered_prices=list(filter(lambda x : x>300,discounted_prices))
print(f"The filtered price is: \n {filtered_prices}" )
```

### Task 7 :- Menu using Functions
#### using function to calculate to add price to the list, return average price and return maximum price and it should be in loop like structure continuously asking user

```python
## The functions:-

def add_price(list,price):
    list.append(price)
    return list

def average_price(list):
    return round((sum(list)/len(list)),2)

def maximum_price(list):
    return max(list)
```

```python
## The menu to control this
price_list=[]

message=" 1. Add Price \n 2. Show Average Price \n 3. Show Maximum Price \n q. Quit"


while True:
    print(message)
    opt=input("Enter the value:")

    ## Adding the item and taking care of invalid value
    if(opt=='1'):
        try:
            price=int(input("Enter the price:"))
            add_price(price_list,price)
            print(f"The item have been added of value {price}")
        except:
           print("Invalid Price is enter")
           continue

    ## Average price if any item is present in list
    elif(opt=='2'):
        if(len(price_list)):
            print(f"The average price of the item present in the list is:{average_price(price_list)}\n")
        else: "The list is empty."

    ## Maximum item if any item is present in list
    elif(opt=='3'):
        if(len(price_list)):
            print(f"The maximum value item present in the list is :{maximum_price(price_list)}\n")
        else: "The list is empty."

    ## Quit 
    elif(opt=='q'):
        print(f"Exiting the menu with list values {sum(price_list)}")
        break

    else:
        print(f"Invalid option please try again")
```
